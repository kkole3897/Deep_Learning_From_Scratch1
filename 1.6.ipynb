# 1.6 MATPLOTLIB

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x= np.arange(0, 6 , 0.1)
y = np.sin(x)
```

```python
plt.plot(x,y)
plt.show()
```

```python
#1.6.2 pyplot의 기능
```

```python
x= np.arange(0, 6, 0.1)
y1= np.sin(x)
y2= np.cos(x)
```

```python
plt.plot(x,y1, label='sin')
plt.plot(x, y2, linestyle='--', label = 'cos')
plt.xlabel('x')
plt.ylabel('y')
plt.title('sin&cos')
plt.legend()
plt.show()
```
